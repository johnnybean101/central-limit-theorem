```python
# Import Libraries
##################################################################################
import numpy as np
from numpy.random import shuffle
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
##################################################################################
```

```python
# Define Functions
##################################################################################
# plot probability distribution for a given discrete random variable
def plot_pDist(x, p_x, p_mean):
    fig, ax = plt.subplots(figsize=(10,5))
    ax.bar(x, p_x, color='steelblue', alpha=0.7)
    ax.set_xlabel('x', fontsize=15)
    ax.set_ylabel('p(x)', fontsize=15)
    ax.set_title('Probability Distribution of Random Variable x', fontsize=15)
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
    plt.axvline(x=p_mean, color='indianred', linestyle='--')
##################################################################################
# return sample means with sample size n
def get_sample_mean(x, p_x, n, n_s):
    sample_means = np.zeros(n_s)
    for index, value in enumerate(sample_means):
        sample_means[index] = np.average(np.random.choice(x, n, p=p_x))
    return sample_means
##################################################################################
# plot sampling distribution for a given discrete random variable
def plot_sDist(sample_means, s_mean):
    fig, ax = plt.subplots(figsize=(10,5))
    ax.hist(sample_means, bins=15, color='steelblue', alpha=0.7)
    ax.set_xlabel('Sample Mean', fontsize=15)
    ax.set_ylabel('Frequency', fontsize=15)
    ax.set_title('Sampling Distribution of Random Variable x', fontsize=15)
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
    plt.axvline(x=s_mean, color='indianred', linestyle='--')
##################################################################################
# plot sample quantiles against quantiles from a normal distribution
def qq_plot(sample_means):
    fig, ax = plt.subplots(figsize=(7.5,5))
    qqplot(sample_means, line='s', ax=ax, alpha=0.5, color='steelblue')
    ax.set_xlabel('Theoretical Quantiles', fontsize=15)
    ax.set_ylabel('Sample Quantiles', fontsize=15)
    ax.set_title('Quantile-Quantile Plot', fontsize=15)
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
##################################################################################
```

```python
# Define Probability Distribution #
##################################################################################
# set n_x possible discrete values and their corresponding probabilities p_x
n_x = 5 
x = np.arange(n_x+1)
p_x = x/np.sum(x)
shuffle(p_x)
##################################################################################
# calculate population mean, population variance, and population standard deviation
p_mean = np.sum(np.multiply(x, p_x))
p_var = np.sum(np.multiply(np.square(x - p_mean), p_x))
p_std = np.sqrt(p_var)
print('Population Mean: ' + str(round(p_mean, 2)))
print('Population Variance: ' + str(round(p_var, 2)))
print('Population Standard Deviation: ' + str(round(p_std, 2)))
##################################################################################
# plot probability distribution
plot_pDist(x, p_x, p_mean)
```

```python
# Get Sampling Distribution #
##################################################################################
# set sample size n and number of samples n_s
n = 20
n_s = 10000
##################################################################################
# calculate n_s sample means
sample_means = get_sample_mean(x, p_x, n, n_s)
##################################################################################
# calculate population mean, population variance, and population standard deviation
s_mean = np.average(sample_means)
s_var = np.var(sample_means)
s_std = np.std(sample_means)
print('Sample Mean: ' + str(round(s_mean, 2)))
print('Sample Variance: ' + str(round(s_var, 2)))
print('Sample Standard Deviation: ' + str(round(s_std, 2)))
##################################################################################
# plot sampling distribution
plot_sDist(sample_means, s_mean)
# plot quantile-quantile 
qq_plot(sample_means)
```
